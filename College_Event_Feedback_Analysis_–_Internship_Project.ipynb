```python
# Data manipulation

import pandas as pd
import numpy as np

# Visualization

import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# NLP / Sentiment Analysis

from textblob import TextBlob


# Display settings

pd.set_option('display.max_columns',None)
sns.set_style('whitegrid')
```

```python
# Codes for downloading the dataset

import pandas as pd
import random
from datetime import datetime, timedelta

# !pip install textblob

from textblob import TextBlob

data = []

# Technical Event Map

events = {'Tech Fest 2023': 'Fest', 'AI Workshop': 'Workshop', 'Hackathon': 'Hackathon', 'Cloud Seminar': 'Seminar'}

for i in range(750):
    e_name = random.choice(list(events.keys()))
    # Rating Weights: 5(50%), 4(30%), 3(10%), 2(5%), 1(5%)
    rating = random.choices([5, 4, 3, 2, 1], weights=[50, 30, 10, 5, 5])[0]

    # Generate active text
    if rating >= 4:
        comment = f"The {e_name} was an amazing and highly informative experience for all students."
    elif rating == 3:
        comment = f"The {e_name} was decent, but it felt a little bit repetitive in some sections."
    else:
        comment = f"I hated the {e_name}, it was terrible and a complete waste of time today."

    data.append([
        (datetime(2023, 1, 1) + timedelta(days=random.randint(0, 360))).strftime('%Y-%m-%d %H:%M:%S'),
        f"STU{2000+i}", e_name, events[e_name], random.choice(['CS', 'IT', 'MECH', 'CIVIL']),
        rating, comment, random.randint(4, 5) if rating > 3 else random.randint(1, 3),
        random.randint(4, 5) if rating > 3 else random.randint(1, 3), 1, random.randint(1, 5)
    ])

df = pd.DataFrame(data, columns=['Timestamp', 'Student_ID', 'Event_Name', 'Event_Type', 'Department', 'Rating', 'Feedback_Comment', 'Subject_Knowledge_Score', 'Concept_Clarity_Score', 'Use_of_PPT', 'Assignment_Difficulty_Level'])
df.to_csv('College_Event_Feedback_Master_Final_Corrected.csv', index=False)
print("SUCCESS: Final 100% correct dataset is ready!")
```

```python
# Load dataset

df = pd.read_csv('/content/College_Event_Feedback_Master_Final_Corrected.csv')

# View first 5 rows

df.head(15)
```

```python
# Dataset structure

df.info()

# Statistical summary

df.describe()

# Check missing values

df.isnull().sum()
```

```python
# Convert numeric columns safely

numeric_cols = ['Rating','Subject_Knowledge_Score','Concept_Clarity_Score','Use_of_PPT','Assignment_Difficulty_Level']


df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce')

# Fill numeric missing values with column mean

df[numeric_cols]=df[numeric_cols].fillna(df[numeric_cols].mean())

# Fill missing feedback text

df['Feedback_Comment']=df['Feedback_Comment'].fillna('')

# Standardize categorical columns

df['Event_Type']=df['Event_Type'].astype(str).str.title()
df['Department']=df['Department'].astype(str).str.title()

df.head()
```

```python
# Rating Distribution

plt.figure(figsize=(8,5))
sns.countplot(x='Rating',data=df)
plt.title('Distribution of Event Ratings')
plt.xlabel('Rating (1-5)')
plt.ylabel('Count')
plt.show()
```

```python
# Average Scores by Event Type

avg_scores=df.groupby('Event_Type')[numeric_cols].mean()

avg_scores
```

```python
# Top 3 Events by Rating

top_events=df.groupby('Event_Name')['Rating'].mean().sort_values(ascending=False).head(3)
top_events
```

```python
# Correlation Heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation between Evaluation Scores')
plt.show()
```

```python
# Sentiment Analysis

def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Feedback_Comment'].apply(get_sentiment)
df['Sentiment'].value_counts()
```

```python
# Sentiment Visualization

sentiment_counts=df['Sentiment'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    sentiment_counts,
    labels=sentiment_counts.index,
    autopct='%1.1f%%',
    colors=['green','grey','red'])

plt.title('Overall Feedback Sentiment')
plt.show()
```

```python
# Word Cloud

text= ' '.join(df['Feedback_Comment'])

wordcloud=WordCloud(

                    width=800,
                    height=400,
                    background_color='white'
).generate(text)


plt.figure(figsize=(15,6))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Words in Student Feedback')
plt.show()
```

```python
# Department-wise Analysis

dept_rating=df.groupby('Department')['Rating'].mean().sort_values(ascending=False)
dept_rating
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=dept_rating.index,y=dept_rating.values)
plt.xticks(rotation=45)
plt.title('Average Event Rating by Department')
plt.ylabel('Average Rating')
plt.show()
```
